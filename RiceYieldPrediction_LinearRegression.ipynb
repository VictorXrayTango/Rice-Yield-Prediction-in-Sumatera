```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/Data_Tanaman_Padi_Sumatera_version_1.csv')

print(data.head())
```

```python
x = data[['Luas Panen','Curah hujan','Kelembapan','Suhu rata-rata']]
y = data[['Produksi']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ", mlr.intercept_)
print("Coefficients:")


for var, coeff in zip(list(x), mlr.coef_[0]):
  print(f'Variable {var} : {coeff}')
```

```python
#Prediction of test set
y_pred_mlr = mlr.predict(x_test)
#Predicted values
print(f"Prediction for test set: \n{y_pred_mlr}")
```

```python
#Actual value and the predicted value

y_test = np.array(y_test)
mlr_diff = pd.DataFrame({'Actual value': y_test[0], 'Predicted value': y_pred_mlr[0]})

mlr_diff.head()
```

```python
#Model Evaluation
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.4f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
